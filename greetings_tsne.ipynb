```python
from main import generate_bow
# from tsne import tsne
from sklearn.manifold import TSNE
import pandas as pd
```

```python
bag_matrix, allsentences = generate_bow()
tsne_values = TSNE(n_components=2).fit_transform(bag_matrix)
```

```python
df = pd.DataFrame()
df['x'] = tsne_values[:,0]
df['y'] = tsne_values[:,1]
df['text'] = allsentences
```
